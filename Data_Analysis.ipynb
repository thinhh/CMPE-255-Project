```python
new_df = pd.read_csv('LA_Auto_Theft.csv')
new_df.head(3)
```

```python
# Let's remove columns that have too many missing values

# List of columns to drop
columns_to_drop = ['Mocodes', 'Vict Sex', 'Vict Descent', 'Weapon Used Cd', 'Weapon Desc', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street']

# Drop the specified columns
new_df = new_df.drop(columns=columns_to_drop)

new_df.head(3)
```

```python
# List of columns to strip spaces from
columns_to_strip = ['AREA NAME', 'Crm Cd Desc', 'Premis Desc', 'Status Desc', 'LOCATION']

# Apply str.strip() to specified columns
new_df[columns_to_strip] = new_df[columns_to_strip].apply(lambda x: x.str.strip() if x.dtype == "O" else x)

new_df[new_df['LOCATION'] == new_df.iloc[0]['LOCATION']]
```

```python
# As visible in the above results let's replace the missing values in 'Premis Cd' and 'Premis Desc' with the corresponding value of the row
# with the same 'LOCATION'

cols_to_replace = ['Premis Cd', 'Premis Desc']

new_df[cols_to_replace] = new_df.groupby('LOCATION')[cols_to_replace].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else x.mean()))

new_df.isna().sum()
```

```python
# Still there are missing values

new_df[new_df['Premis Cd'].isna()]
```

```python
# We will drop that row

new_df = new_df.drop(new_df[new_df['Premis Cd'].isna()].index)

new_df.isna().sum()
```

```python
# Remove unnecessary spaces
new_df['LOCATION'] = new_df['LOCATION'].str.replace('\s+', ' ', regex=True).str.strip()
# uniqueAddress[0:10]

# Print unique locations
uniqueAddress = new_df['LOCATION'].unique()
print("Total Rows: ", uniqueAddress.size)
uniqueAddress[0:11]
# for address in uniqueAddress:
#     print(address,"\n")
```

```python
#Bring date to proper format

def keep_date(text):
    return text.split()[0]

new_df['DATE OCC'] = new_df['DATE OCC'].apply(lambda x: keep_date(x))
new_df['Date Rptd'] = new_df['Date Rptd'].apply(lambda x: keep_date(x))

new_df.head(3)
```

```python
# Convert date-related columns to datetime objects
new_df['Date Rptd'] = pd.to_datetime(new_df['Date Rptd'], errors='coerce')
new_df['DATE OCC'] = pd.to_datetime(new_df['DATE OCC'], errors='coerce')

# Extract the first two characters of 'TIME OCC' and convert to int
new_df['Hour'] = pd.to_numeric(new_df['TIME OCC'].astype(str).str[:2], errors='coerce', downcast='integer')

# Drop rows with missing dates
new_df = new_df.dropna(subset=['Date Rptd', 'DATE OCC', 'Hour'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract day of the week, month, and hour
new_df['Day_of_Week'] = new_df['DATE OCC'].dt.day_name()
new_df['Month'] = new_df['DATE OCC'].dt.month

# Plot crime occurrences by day of the week, month, and hour
plt.figure(figsize=(15, 5))
sns.countplot(x='Day_of_Week', data=new_df, order=new_df['Day_of_Week'].value_counts().index)
plt.title('Auto Theft Crimes by Day of the Week')

plt.figure(figsize=(15, 5))
sns.countplot(x='Month', data=new_df, order=new_df['Month'].value_counts().index)
plt.title('Auto Theft Crimes by Month')

plt.figure(figsize=(15, 5))
sns.countplot(x='Hour', data=new_df, order=new_df['Hour'].value_counts().index)
plt.title('Auto Theft Crimes by Hour')

plt.show()
```

```python
# Plotting crime occurrences on a map
plt.figure(figsize=(10, 10))
sns.scatterplot(x='LON', y='LAT', data=new_df, hue='AREA NAME', legend='full', s=20)
plt.title('Spatial Distribution of Auto Theft Crimes')
plt.show()
```

```python
# We can see some rows with 0 for LAT and LON

locations_to_check = new_df[(new_df['LON'] == 0) | (new_df['LAT'] == 0)]['LOCATION']
print("Location with incorrect co-ordinates: ",locations_to_check,"\n")

for location in locations_to_check:
    print(new_df[new_df['LOCATION'] == location]['LON'])
    print(new_df[new_df['LOCATION'] == location]['LAT'],"\n")
```

```python
# We can not use LOCATION to replace 0 values because all have 0 values in LAT and LON
# Therefore we will drop them

new_df = new_df.drop(new_df[(new_df['LON'] == 0) | (new_df['LAT'] == 0)].index)

locations_to_check = new_df[(new_df['LON'] == 0) | (new_df['LAT'] == 0)]['LOCATION']
print("Location with incorrect co-ordinates: ",locations_to_check,"\n")
```

```python
# Plotting crime occurrences on a map
plt.figure(figsize=(10, 10))
sns.scatterplot(x='LON', y='LAT', data=new_df, hue='AREA NAME', legend='full', s=20)
plt.title('Spatial Distribution of Auto Theft Crimes')
plt.show()
```

```python
# Only keep rows with LON between -110 to -120 and LAT between 32.5 to 35

# Specify the conditions for filtering
lon_condition = (new_df['LON'] >= -120) & (new_df['LON'] <= -110)
lat_condition = (new_df['LAT'] >= 32.5) & (new_df['LAT'] <= 35)

# Keep only the rows that satisfy both conditions
new_df = new_df[lon_condition & lat_condition]
```

```python
# Plotting crime occurrences on a map
plt.figure(figsize=(10, 10))
sns.scatterplot(x='LON', y='LAT', data=new_df, hue='AREA NAME', legend='full', s=20, alpha=0.6, palette='bright')

# Add a new column with the crime frequency for each row
new_df['CrimeFrequency'] = new_df.groupby('AREA NAME')['AREA NAME'].transform('count')

sns.kdeplot(x='LON', y='LAT', data=new_df, cmap='viridis', fill=True, alpha=0.3, weights=new_df['CrimeFrequency'])
plt.title('Spatial Distribution of Auto Theft Crimes with Heatmap')
plt.show()
```

```python
# Plotting histogram for frequency of crime by area
plt.figure(figsize=(12, 6))
sns.histplot(data=new_df, x='AREA NAME', bins=len(new_df['AREA NAME'].unique()), palette='viridis')
plt.title('Histogram of Crime Frequency by Area')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Crime frequency by area by hour

crime_frequency = new_df.groupby(['AREA NAME', 'Hour']).size().reset_index(name='crime_count')
print(crime_frequency.head(3), "\n")

crime_pivot = crime_frequency.pivot(index='Hour', columns='AREA NAME', values='crime_count').fillna(0)

plt.figure(figsize=(12, 8))
sns.heatmap(crime_pivot, cmap='Blues', annot=True, fmt='g', linewidths=.5)
plt.title('Crime Frequency by Area and Hour')
plt.xlabel('Area')
plt.ylabel('Hour of the Day')
plt.show()
```

```python
new_df['DATE OCC'] = pd.to_datetime(new_df['DATE OCC'])

years = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
filtered_dates_list = []
# Extract all dates for a particular year
for target_year in years:
    filtered_dates_list.extend([new_df['DATE OCC'].apply(lambda date: date if date.year == target_year else pd.NaT).dropna()])
filtered_dates_list
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

for filtered_dates in filtered_dates_list:
    # Plotting histogram for filtered dates
    plt.figure(figsize=(10, 6))
    plt.hist(filtered_dates, bins=335, edgecolor='black')
    plt.title(f'Distribution of Filtered Dates')
    plt.xlabel('Date')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.ylabel('Frequency')
    plt.tight_layout()  # Optimize layout
    plt.show()
```

```python
# Dates with higher crimes
#!pip install plotly
import matplotlib.pyplot as plt

def plot_by_day(threshold,date_counts):
# Filter dates with frequency greater than threshold
    dates_over_threshold = date_counts[date_counts > threshold].index
    
    # Get day of the week for each date
    day_of_week = dates_over_threshold.to_series().dt.day_name()
    
    # Print the result
    days_of_week_df = pd.DataFrame({'Date': dates_over_threshold, 'Day of the Week': day_of_week})
    
    # Plotting histogram for filtered dates
    plt.figure(figsize=(10, 8))
    plt.hist(days_of_week_df['Day of the Week'], bins=7, edgecolor='black')
    plt.title('Distribution of Filtered Dates')
    plt.xlabel('Day')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.ylabel('Frequency')
    plt.tight_layout()  # Optimize layout
    plt.show()
```

```python

for filtered_dates in filtered_dates_list:
    # Get frequency of each date
    date_counts = filtered_dates.value_counts().sort_index()
    
    x_values = date_counts.index
    y_values = date_counts.values
    
    # Create a scatter plot
    plt.scatter(x_values, y_values, label='Data Points')
    
    # Add labels and title
    plt.xlabel('Dates')
    plt.ylabel('Frequency')
    plt.title('Scatter Plot')
    plt.legend()
    
    # Show the plot
    plt.show()

    plot_by_day(55, date_counts)
    # date_counts.index
    # date_counts.values
```
